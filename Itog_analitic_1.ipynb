```python
# Разделение данных на контрольную и тестовую группы
gate_A_group = df[df['ab_group'] == 'A']
gate_B_group = df[df['ab_group'] == 'B']
```

```python
gate_A_group
```

```python
gate_B_group
```

```python
# Проведение теста Манна-Уитни
statistic, p_value = mannwhitneyu(gate_B_group['revenue'], gate_A_group['revenue'])
```

```python
print("Статистика теста:", statistic)
print("p-значение:", p_value)
```

```python
# Рассчет конверсии в группах
conversion_gate_A_group = gate_A_group['revenue'].mean()
conversion_gate_B_group = gate_B_group['revenue'].mean()
```

```python
print("Конверсия gate_A_group:", conversion_gate_A_group)
print("Конверсия gate_B_group:", conversion_gate_B_group)
```

Мы провели тест Манна-Уитни и выполнили расчет конверсии в группах беря в качестве метрик revenue (доход) и видим что конверсия  выше для группы gate_А_group.
